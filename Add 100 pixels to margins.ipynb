```python


def add_padding_to_image(image_path, output_path, padding=100):
    # Open an image
    img = Image.open(image_path)
    
    # Get the current size of the image
    width, height = img.size
    
    # Create a new image with the new size (adding padding to each side)
    new_img = Image.new("RGB", (width + 2 * padding, height + 2 * padding), (255, 255, 255))
    
    # Paste the original image in the center of the new image
    new_img.paste(img, (padding, padding))
    
    # Save the new image
    new_img.save(output_path)

def process_images_in_folder(folder_path, output_folder, padding=100):
    # Check if output folder exists, if not create it
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Iterate over all files in the folder
    for filename in os.listdir(folder_path):
        # Get file extension
        file_extension = filename.split('.')[-1].lower()
        
        # Process only supported image types
        if file_extension in ['jpeg', 'jpg', 'png', 'tif']:
            image_path = os.path.join(folder_path, filename)
            output_path = os.path.join(output_folder, filename)
            add_padding_to_image(image_path, output_path, padding)
            print(f"Processed {filename}")

process_images_in_folder(folder_path, output_folder)
```
