```python
# Step 0. Load the dependencies
import cv2
import os
import glob
```

```python
# Step 1. Path to the folder containing images
input_folder = "/Users/davidnguyen/Documents/Dave Personal Stuff/NLP/Image Analysis Workshop/Student Files/Maida/Shahzad2022_Fig2A_processed/Step 2_Shahzad2022_Fig2A Indiv Masks_trim1px"

# Step 2 and on:
output_folder = os.path.join(input_folder, "enlarged")

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Get all image files in the folder (jpg, png, etc.)
image_files = glob.glob(os.path.join(input_folder, "*.*"))

for image_path in image_files:
    # Read the image
    img = cv2.imread(image_path)

    if img is None:
        print(f"Skipping {os.path.basename(image_path)} - not a valid image")
        continue

    # Enlarge by factor of 3
    enlarged_img = cv2.resize(img, None, fx=3, fy=3, interpolation=cv2.INTER_CUBIC)

    # Get filename and extension
    base_name = os.path.basename(image_path)
    name, ext = os.path.splitext(base_name)

    # Save with "_mag3x" suffix in output folder
    output_path = os.path.join(output_folder, f"{name}_mag3x{ext}")
    cv2.imwrite(output_path, enlarged_img)

    print(f"Saved: {output_path}")
```
