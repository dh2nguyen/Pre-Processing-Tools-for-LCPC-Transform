```python
import cv2
import os
```

```python

# Set input and output folders
input_folder = "/Users/davidnguyen/Documents/Dave Personal Stuff/TSG Lab/Projects/Human Rib Chest X-ray/Pure Shape_30s/PureShape - Women 30s/Step 8 Thicken blue lines"  # Change this to your folder
output_folder = "/Users/davidnguyen/Documents/Dave Personal Stuff/TSG Lab/Projects/Human Rib Chest X-ray/Pure Shape_30s/PureShape - Women 30s/Step 9 Resize 400 wide"  # Change this to where resized images should be saved
```

```python

# Ensure output folder exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Target width
TARGET_WIDTH = 400

# Process each image in the folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.jpeg', '.jpg', '.png', '.tif')):
        img_path = os.path.join(input_folder, filename)
        img = cv2.imread(img_path)

        if img is None:
            print(f"Skipping {filename}: Cannot read file")
            continue

        # Get original dimensions
        h, w = img.shape[:2]

        # Calculate new height while maintaining aspect ratio
        aspect_ratio = h / w
        new_height = int(TARGET_WIDTH * aspect_ratio)

        # Resize image
        resized_img = cv2.resize(img, (TARGET_WIDTH, new_height), interpolation=cv2.INTER_AREA)

        # Save resized image
        save_path = os.path.join(output_folder, filename)
        cv2.imwrite(save_path, resized_img)
        print(f"Resized and saved: {save_path}")

print("Processing complete!")
```
