```python
# Rotate images in a folder 90 degrees counter-clockwise
```

```python
import cv2
import os
```

```python
# Define the input and output folder paths manually
input_folder = r"/Users/davidnguyen/Documents/Dave Personal Stuff/TSG Lab/Projects/Human Rib Chest X-ray/Pure Shape_30s/PureShape - Women 30s/Step 5 Rotate 90 Degree Clockwise_WOMEN"
output_folder = r"/Users/davidnguyen/Documents/Dave Personal Stuff/TSG Lab/Projects/Human Rib Chest X-ray/Pure Shape_30s/PureShape - Women 30s/Step 6 Rotate 90 CCW"
```

```python
# Ensure the input folder exists
if not os.path.exists(input_folder):
    print("Error: The input folder does not exist.")
    exit()

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Get a list of all files in the input folder
files = os.listdir(input_folder)

# Process each file
for file_name in files:
    input_file_path = os.path.join(input_folder, file_name)

    # Check if it's an image
    if file_name.lower().endswith((".png", ".jpg", ".jpeg", ".tif")):
        try:
            # Read the image
            img = cv2.imread(input_file_path)

            if img is None:
                print(f"Skipping {file_name}: Unable to load image.")
                continue

            # Rotate 90 degrees clockwise
            rotated_img = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

            # Save the rotated image with a new filename
            name, ext = os.path.splitext(file_name)
            rotated_file_name = f"{name}_rot90cw{ext}"
            rotated_file_path = os.path.join(output_folder, rotated_file_name)
            cv2.imwrite(rotated_file_path, rotated_img)

            print(f"Rotated and saved: {rotated_file_path}")

        except Exception as e:
            print(f"Error processing {file_name}: {e}")

print("Rotation process completed.")
```
